# Notebook 04 — SARIMA

Este notebook implementa el modelo SARIMA para la predicción de la
huella de carbono operacional del sistema eléctrico español con
resolución de 15 minutos.

SARIMA extiende ARIMA añadiendo una componente estacional explícita,
lo que permite capturar el patrón diario de la serie sin necesidad
de un gran número de rezagos.

**Orden seleccionado:** SARIMA(2,1,0)(1,1,0)[96]

**Evaluación:** walk-forward con 12 fits distribuidos por 2024,
misma configuración que el resto de modelos.

**Horizontes:** 48h (192 pasos) y 72h (288 pasos).

```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from dataclasses import dataclass
from sklearn.metrics import mean_absolute_error, mean_squared_error
from statsmodels.tsa.statespace.sarimax import SARIMAX

warnings.filterwarnings("ignore")
plt.style.use("seaborn-v0_8")
%matplotlib inline

BASE_DIR    = Path("/home/ubuntu/TFM")
DATA_DIR    = BASE_DIR / "notebooks/data_processed"
RESULTS_DIR = BASE_DIR / "results"
RESULTS_DIR.mkdir(exist_ok=True)

print("Librerías cargadas OK")
```

## 1. Carga de datos

```python
y_train = pd.read_parquet(DATA_DIR / "train_2022_2023.parquet")["y"].astype(float)
y_val   = pd.read_parquet(DATA_DIR / "val_2024.parquet")["y"].astype(float)
y_test  = pd.read_parquet(DATA_DIR / "test_2025.parquet")["y"].astype(float)

y_train.index = pd.to_datetime(y_train.index)
y_val.index   = pd.to_datetime(y_val.index)
y_test.index  = pd.to_datetime(y_test.index)

print("Train:", y_train.shape, "|", y_train.index.min(), "->", y_train.index.max())
print("Val:  ", y_val.shape,   "|", y_val.index.min(),   "->", y_val.index.max())
print("Test: ", y_test.shape,  "|", y_test.index.min(),  "->", y_test.index.max())
```

## 2. Configuración temporal y métricas

```python
FREQ_MIN        = 15
STEPS_PER_HOUR  = 60 // FREQ_MIN
SEASONAL_PERIOD = 24 * STEPS_PER_HOUR  # 96

HORIZONS = {
    "48h": 48 * STEPS_PER_HOUR,  # 192 pasos
    "72h": 72 * STEPS_PER_HOUR,  # 288 pasos
}

@dataclass
class WFConfig:
    step: int
    min_history: int
    max_fits: int

def compute_metrics(y_true, y_pred):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    return {"MAE": mae, "RMSE": rmse}

print("Periodo estacional:", SEASONAL_PERIOD)
print("Horizontes:", HORIZONS)
```

## 3. Selección del orden SARIMA — auto_arima

Se usa auto_arima con seasonal=True y period=96 sobre los últimos
7 días del conjunto de entrenamiento para reducir el coste
computacional de la búsqueda.

Se limita el espacio de búsqueda a órdenes pequeños dado el
elevado coste computacional de SARIMA con period=96.

```python
from pmdarima import auto_arima

# 7 días para que sea viable computacionalmente
sample = y_train.iloc[-7 * SEASONAL_PERIOD:]

print("Buscando orden óptimo con auto_arima seasonal=True...")

auto_model = auto_arima(
    sample.values,
    start_p=0, max_p=3,
    start_q=0, max_q=1,
    d=None,
    start_P=0, max_P=2,
    start_Q=0, max_Q=1,
    D=None,
    seasonal=True,
    m=96,
    information_criterion="aic",
    stepwise=True,
    suppress_warnings=True,
    error_action="ignore",
    verbose=True,    
    trace=True       
)

sarima_order          = auto_model.order
sarima_seasonal_order = auto_model.seasonal_order
print("\nOrden óptimo:", sarima_order)
print("Orden estacional:", sarima_seasonal_order)
print("AIC:", round(auto_model.aic(), 2))
print(auto_model.summary())
```

## 4. Interpretación del orden seleccionado

auto_arima confirma **SARIMA(1,1,0)(1,0,0)[96]** como orden óptimo:

**Parte no estacional:**
- d=1 -> la serie necesita diferenciación para ser estacionaria
- p=1 -> dependencia con el instante anterior
- q=0 -> no se necesitan términos de media móvil

**Parte estacional:**
- D=0 -> no necesita diferenciación estacional
- P=1 -> dependencia con el mismo instante hace 96 pasos (24h antes)
- Q=0 -> sin media móvil estacional

**Coeficientes:**
- ar.L1=0.54 -> dependencia positiva fuerte con el instante anterior
- ar.S.L96=0.27 -> dependencia positiva con hace exactamente 24h

Todos los coeficientes son estadísticamente significativos (p<0.001).

**Diagnóstico de residuos:**
- Ljung-Box p=0.67 -> residuos no autocorrelacionados 
- Jarque-Bera p=0.00 -> residuos no normales (curtosis=8.92)

```python
cfg_sarima = WFConfig(
    step=28 * SEASONAL_PERIOD,
    min_history=30 * SEASONAL_PERIOD,
    max_fits=12
)

def walk_forward_splits(series, horizons, cfg):
    n     = len(series)
    max_h = max(horizons.values())
    points = list(range(cfg.min_history, n - max_h, cfg.step))
    points = points[:cfg.max_fits]
    for t0 in points:
        train_part = series.iloc[:t0]
        tests = {k: series.iloc[t0:t0 + h] for k, h in horizons.items()}
        yield train_part, tests

def summarize(df, model_name):
    out = (
        df.groupby("horizon")[["MAE", "RMSE"]]
        .mean()
        .reset_index()
    )
    out.insert(0, "model", model_name)
    return out

def compute_metrics(y_true, y_pred):
    mae  = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    return {"MAE": mae, "RMSE": rmse}

print("Walk-forward configurado OK")
```

## 5. Modelo SARIMA

SARIMA(p,d,q)(P,D,Q)[s] extiende ARIMA con una componente estacional:
- **(p,d,q)**: parte no estacional — igual que ARIMA
- **(P,D,Q)**: parte estacional — captura patrones periódicos
- **s=96**: periodo estacional — un día completo a resolución 15min

El coste computacional es elevado porque la representación en espacio
de estados con s=96 genera matrices de gran dimensión.
Por ello se limita maxiter=5.

```python
def forecast_sarima(train, h, order, seasonal_order):
    model = SARIMAX(
        train.values,
        order=order,
        seasonal_order=seasonal_order,
        enforce_stationarity=False,
        enforce_invertibility=False
    )
    res = model.fit(
        disp=False,
        maxiter=5,
        low_memory=True,
        cov_type="none"
    )
    return res.forecast(steps=h)

print("forecast_sarima definido OK")
```

## 6. Evaluación SARIMA con walk-forward

```python
def evaluate_sarima(y_train, y_val, horizons, cfg, order, seasonal_order, verbose=True):
    rows        = []
    preds_store = []
    y_full  = pd.concat([y_train, y_val]).sort_index()
    n_train = len(y_train)
    splits  = list(walk_forward_splits(y_val, horizons, cfg))
    total   = len(splits)

    for i, (train_part, tests) in enumerate(splits, 1):
        if verbose:
            print(f"Fit {i}/{total} — entrenando...")
        cutoff_idx   = n_train + len(train_part)
        cutoff_time  = y_full.index[cutoff_idx - 1]
        train_window = y_full.iloc[:cutoff_idx]

        fit_preds = {"cutoff_time": cutoff_time, "train_window": train_window}
        for name, test in tests.items():
            pred = forecast_sarima(train_window, len(test), order, seasonal_order)
            m    = compute_metrics(test.values, pred)
            rows.append({"horizon": name, "MAE": m["MAE"], "RMSE": m["RMSE"]})
            fit_preds[name] = {"pred": pred, "test": test}
        preds_store.append(fit_preds)

    return pd.DataFrame(rows), preds_store

print("evaluate_sarima definido OK")
```

```python
sarima_order          = (1, 1, 0)
sarima_seasonal_order = (1, 0, 0, 96)

print(f"Evaluando SARIMA{sarima_order}x{sarima_seasonal_order}...")
sarima_val, preds_sarima = evaluate_sarima(
    y_train, y_val,
    HORIZONS, cfg_sarima,
    sarima_order,
    sarima_seasonal_order,
    verbose=True
)

sarima_summary = summarize(sarima_val, f"SARIMA{sarima_order}x{sarima_seasonal_order}")
sarima_summary.to_csv(RESULTS_DIR / "sarima_summary.csv", index=False)
print(sarima_summary)
```

## 6. Visualización SARIMA — predicciones vs real

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("SARIMA(1,1,0)(1,0,0)[96] — Predicciones vs Real", fontsize=14)

for idx, split_idx in enumerate([0, 3, 6, 9]):
    ax = axes[idx // 2][idx % 2]
    fit         = preds_sarima[split_idx]
    cutoff_time = fit["cutoff_time"]
    train_w     = fit["train_window"]
    test_48     = fit["48h"]["test"]
    pred_48     = fit["48h"]["pred"]

    history = train_w.iloc[-4 * SEASONAL_PERIOD:]
    ax.plot(history.index, history.values,
            color="gray", label="Historia", linewidth=0.8)
    ax.plot(test_48.index, test_48.values,
            color="blue", label="Real", linewidth=1.2)
    ax.plot(test_48.index, pred_48,
            color="red", linestyle="--", label="SARIMA 48h", linewidth=1.2)
    ax.axvline(x=cutoff_time, color="black", linestyle=":", linewidth=1.0)
    ax.set_title(f"Fit {split_idx+1} — {cutoff_time.strftime('%B %Y')}")
    ax.set_ylabel("gCO2eq/kWh")
    ax.legend(fontsize=8)
    ax.tick_params(axis="x", rotation=30)

plt.tight_layout()
plt.savefig("/home/ubuntu/TFM/figures/sarima_predicciones.png", dpi=150, bbox_inches="tight")
plt.show()
```

## 7. Interpretación de resultados SARIMA

SARIMA(1,1,0)(1,0,0)[96] obtiene un MAE de 25.20 gCO2eq/kWh a 48h
y 29.04 gCO2eq/kWh a 72h.

A pesar de incorporar una componente estacional explícita con
period=96, SARIMA no supera al Naive_seasonal. Hay dos causas:

**1. Órdenes bajos p=1, P=1:**
El modelo solo recuerda 1 rezago no estacional (15 minutos) y
1 rezago estacional (24 horas). Con horizontes de 48h y 72h
la predicción colapsa rápidamente a la media, igual que ARIMA.

**2. Coste computacional:**
SARIMA con period=96 genera aproximadamente 9GB por iteración
debido a la representación en espacio de estados. Esto limita
el número de iteraciones del optimizador y la complejidad
del modelo que se puede ajustar en tiempo razonable.

**Conclusión:**
Ni ARIMA ni SARIMA superan al Naive_seasonal para horizontes
largos. Esto justifica el uso de AutoReg con 576 rezagos —
un modelo que mantiene la eficiencia computacional de la
regresión lineal pero con memoria suficiente para capturar
el patrón diario completo.

```python
# Comparación Naive, ARIMA, SARIMA
naive_arima = pd.read_csv(RESULTS_DIR / "comparacion_naive_arima.csv")
sarima_summary = pd.read_csv(RESULTS_DIR / "sarima_summary.csv")

comparacion = pd.concat([naive_arima, sarima_summary], ignore_index=True)
comparacion["MAE"]  = comparacion["MAE"].round(3)
comparacion["RMSE"] = comparacion["RMSE"].round(3)

comparacion.to_csv(RESULTS_DIR / "comparacion_naive_arima_sarima.csv", index=False)
print(comparacion)
```

## 8. Comparación final 

```python
comparacion = pd.concat([
    naive_arima,
    sarima_summary
], ignore_index=True)

comparacion["MAE"]  = comparacion["MAE"].round(3)
comparacion["RMSE"] = comparacion["RMSE"].round(3)

tabla = comparacion.pivot_table(
    index="model",
    columns="horizon",
    values=["MAE", "RMSE"]
).round(3)

tabla = tabla.sort_values(("MAE", "48h"))

print(tabla)
comparacion.to_csv(RESULTS_DIR / "comparacion_naive_arima_sarima.csv", index=False)
```

## 9. Interpretación de resultados y justificación de AutoReg

| Modelo | MAE 48h | MAE 72h |
|--------|---------|---------|
| Naive_seasonal | 18.30 | 19.66 |
| SARIMA(1,1,0)(1,0,0)[96] | 25.20 | 29.04 |
| ARIMA(5,1,0) | 27.16 | 28.06 |
| Naive_last | 27.91 | 29.78 |

### Conclusiones

Ninguno de los modelos estadísticos clásicos supera al Naive_seasonal,
que se mantiene como baseline principal con MAE 18.30 a 48h.

**ARIMA(5,1,0)** obtiene resultados similares a Naive_last. Con solo
p=5 rezagos (75 minutos de memoria) la predicción colapsa a la media
para horizontes de 48h y 72h, sin capturar el patrón diario.

**SARIMA(1,1,0)(1,0,0)[96]** incorpora una componente estacional
explícita con period=96, lo que le permite capturar dependencias
diarias. Sin embargo tampoco supera al Naive_seasonal. La causa es
doble: los órdenes bajos (p=1, P=1) limitan la memoria del modelo,
y el elevado coste computacional de SARIMA con period=96 impide usar órdenes más altos en tiempo razonable.

### Motivación para AutoReg

Estos resultados motivan el uso de **AutoReg**. AutoReg resuelve los dos problemas
identificados:

1. **Memoria:** usa 576 rezagos (6 días de historia) frente a los
   5 de ARIMA y los 1+1 de SARIMA, capturando dependencias de
   largo alcance incluyendo el patrón diario completo.

2. **Eficiencia computacional:** al ser una regresión lineal directa
   no necesita representación en espacio de estados, eliminando
   el problema del alto coste por iteración de SARIMA.

3. **Features cíclicas:** incorpora el seno y coseno de la hora
   como variables exógenas, codificando explícitamente la
   periodicidad diaria de forma continua y diferenciable.
